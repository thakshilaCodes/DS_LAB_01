```python
import matplotlib.pyplot as plt

exp_data = np.random.exponential(scale=1.0, size=100000)
uni_data = np.random.rand(100000)
norm_data = np.random.randn(100000)

plt.hist(exp_data, bins=200, density=True, histtype='step', color='red', label='Exponential')
plt.hist(uni_data, bins=200, density=True, histtype='step', color='blue', label='Uniform')
plt.hist(norm_data, bins=200, density=True, histtype='step', color='green', label='Normal')

plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.axis([-2, 8, 0, 1.2])
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot of Z = X^2 + Y^2')
plt.show()
```

```python
import pandas as pd
import seaborn as sns

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv

df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

df.head()

features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
stats_df = df[features]

pearson_corr = stats_df.corr(method='pearson')
spearman_corr = stats_df.corr(method='spearman')

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title('Pearson Correlation')

plt.subplot(1,2,2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm')
plt.title('Spearman Correlation')

plt.show()
```
